```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style = 'whitegrid')
```

**Q. 5) Which are Top Selling Courses ?**

```python
df = data.sort_values('num_subscribers', ascending = False).head(5)
```

```python
df
```

```python
new_df = df[['course_title', 'num_subscribers']]
```

```python
collist = df['course_title'].values.tolist()
```

```python
collist
```

```python
subs = df['num_subscribers'].values.tolist()
subs
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x = collist, y = subs, dodge = False, palette = 'autumn')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.title('Top selling courses', weight='bold').set_fontsize('15')
plt.gcf().set_size_inches(12, 9)
plt.show()
```

**Q. 10) What is the Max. Number of Subscribers for Each Level of courses ?**

```python
af = pd.DataFrame(data.groupby('level')['num_subscribers'].max())
```

```python
af
```

```python
max = data.groupby('level').max()
```

```python
max
```

```python
names = ['All Levels','Beginner Level', 'Expert Level', 'Intermediate Level' ]
```

```python
val = max['num_subscribers'].values.tolist()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x = names, y = val, dodge = False, palette = 'Wistia')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.title('Max no of subscribers for each level', weight='bold').set_fontsize('15')
plt.gcf().set_size_inches(12, 9)
plt.show()
```
